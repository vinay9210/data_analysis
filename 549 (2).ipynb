```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Completed_Building_Details_for_the_year_2023-24.csv")
df
```

```python
sns.histplot(df.iloc[:,1], bins=30, kde=True)
plt.title('Histogram of Column 1')
plt.show()
```

```python
sns.countplot(x=df.iloc[:,2])
plt.title('Count of Column 2 Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x=df.iloc[:,3], y=df.iloc[:,4])
plt.title('Scatter Plot between Column 3 and Column 4')
plt.show()
```
